```python
# Crawl Data

filename = 'filename.csv'
search_keyword = 'iphone min_faves:100 lang:en until:2023-05-31 since:2023-05-01'
limit = 800
token = 'token'

!npx --yes tweet-harvest@2.2.8 -o "{filename}" -s "{search_keyword}" -l {limit} --token "{token}"
```

```python
import pandas as pd

# Specify the path to your CSV file
file_path = f"tweets-data/{filename}"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, delimiter=";")

# Display the DataFrame
display(df)
```
